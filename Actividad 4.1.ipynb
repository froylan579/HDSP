```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
muertes = pd.read_csv('def17cont.csv')
```

```python
muertes.head()
```

```python
maximo = muertes['CAUSA_DEF'].value_counts()
print(maximo)
```

```python
maximo1 = maximo.head(3)
```

```python
fig = plt.figure(figsize=(100,100))
maximo1.plot(kind='bar')
plt.xticks(range (3), fontsize=60, rotation='horizontal')
plt.bar(range(3),maximo1, align='center', color=['blue','red','green'])
plt.title("Principales causas de muerte en MÃ©xico durante 2017", fontsize=100)
plt.xlabel("")
plt.ylabel(" ")
plt.yticks(fontsize=60)
```

```python
I219 = muertes[muertes['CAUSA_DEF']=='I219']
E119 = muertes[muertes['CAUSA_DEF']=='E119']
E112 = muertes[muertes['CAUSA_DEF']=='E112']
#print(I219['ENT_REGIS'].value_counts(ascending=True))
#print(E119['ENT_REGIS'].value_counts(ascending=True))
#print(E112['ENT_REGIS'].value_counts(ascending=True))
```
